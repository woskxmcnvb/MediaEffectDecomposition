# Hills

```python
#########################
# Hills transformation definition and demo 
#########################

def HillsTransformation(x, alpha=1, gamma=1): 
    return (x ** gamma) / (x ** gamma + alpha ** gamma)
    

x = np.arange(0, 30, dtype='double')

fig, axs = plt.subplots(3, 1, figsize=(15,15))

gamma = 1
for alpha in [0.1, 0.2, 0.5, 1, 2, 4, 8]:
    axs[0].plot(x, HillsTransformation(x, alpha, gamma), 
                label='alpha:' + str(alpha) + ' gamma:' + str(gamma))

alpha = 5
for gamma in [1, 2, 4, 6]:
    axs[1].plot(x, HillsTransformation(x, alpha, gamma), 
                label='alpha:' + str(alpha) + ' gamma:' + str(gamma))
        


for alpha in [0.1, 0.2, 0.5, 1, 2, 4, 8]:
    for gamma in [1, 2, 4, 6]:
        axs[2].plot(x, HillsTransformation(x, alpha, gamma), 
                    label='alpha:' + str(alpha) + ' gamma:' + str(gamma))

for ax in axs: 
    ax.set_xticks(x)
    ax.legend()
    ax.grid(True)


plt.show()
```

```python
#########################
# Illustration of Hills transformation с заданными параметрами
alpha = 5
gamma = 1.8
#########################


x = np.arange(0, 20, dtype='double')

fig, ax = plt.subplots(1, 1, figsize=(15, 5))

plt.plot(x, HillsTransformation(x, alpha, gamma), 
                label='alpha:' + str(alpha) + ' gamma:' + str(gamma))

ax.set_xticks(x)
ax.legend()
ax.grid(True)

plt.show()
```

# Exponential

```python
def exponential_saturation(x, rate):
    return 1 - np.exp(-rate * x)
```

```python
x = np.arange(0, 30, dtype='double')

fig, axs = plt.subplots(1, 1, figsize=(15,5))

for rate in [0.1, 0.2, 0.5, 1, 2, 4, 8]:
    axs.plot(x, exponential_saturation(x, rate), 
                label='rate: {}'.format(rate))

axs.set_xticks(x)
axs.legend()
axs.grid(True)


plt.show()
```

# Logit

```python
def logistic_function(x, rate, midpoint):
    return 1 / (1 + np.exp(-rate * (x - midpoint)))
```

```python
x = np.arange(0, 30, dtype='double')

fig, axs = plt.subplots(1, 1, figsize=(15,5))

for rate in [0.1, 0.2, 0.5, 1, 2, 4, 8]:
    for midpoint in [1,2,3,4,5,10,20]:
        axs.plot(x, logistic_function(x, rate, midpoint), 
                    label='rate: {}, midpoint: {}'.format(rate, midpoint))

axs.set_xticks(x)
axs.legend()
axs.grid(True)


plt.show()
```
